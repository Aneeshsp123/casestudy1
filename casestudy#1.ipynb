# Qn3 Plot a bar graph of male vs female buyers participated in the sales

```python
Gender=data['Buyer Gender'].value_counts().to_frame().reset_index()
plt.bar(Gender['index'],Gender['Buyer Gender'],color=['blue','red'])
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('count')
```

```python
the bar graph shows the men and women buyer statistics,this shows that there counts are eqaul.
```
